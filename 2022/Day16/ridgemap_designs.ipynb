```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from ridge_map import RidgeMap

from palettable.cartocolors.sequential import Mint_6_r, SunsetDark_7_r, agGrnYl_7, TealGrn_7_r, PinkYl_6, SunsetDark_7
from palettable.cmocean.sequential import Tempo_6_r
```

# Map 1: Hills of Kassel, Germany

```python
#dictionary of places to map
places = {"Bergpark": [51.324465 , 9.407815],
          'Kaufunger Wald': [51.28966810493981, 9.72952075711647],
          'Bilstein': [51.22662533486517, 9.356632350442183] ,
          'Schauenburg': [51.28546330296784, 9.333240973641763 ],
          'Staufenberg': [51.353179365286564, 9.633814465851367] ,
          'Zierenberg': [51.368913510012945, 9.294643247434456],
          'SÃ¶hrewald': [51.21516778434333, 9.588708678316747 ],
          'Reichenbacher\nKalkberge': [51.17115272810074, 9.77203339029025],
          #'Hohekopf':[51.24611298893059, 9.79729281492084]
         }
```

```python
# map boundaries
tr = [51.426149690270406, 9.956495445717728][::-1]
bl = [51.13401963260046, 9.072315365023943][::-1]

#colours
bgcolor = "#111111"
cmap = SunsetDark_7_r.mpl_colormap

#Setup figure
fig,ax = plt.subplots(figsize=(12, 12))
fig.set_facecolor(bgcolor)

#========
#plot elevation lines
rm = RidgeMap(bl + tr, font = "Ubuntu")
values = rm.get_elevation_data(num_lines=120, elevation_pts=600)
ridges = rm.plot_map(values=rm.preprocess(values=values, vertical_ratio=120, water_ntile=0, 
                                 lake_flatness=6), 
            ax=ax, kind='elevation', label=None,
            linewidth=1.5,
            line_color = plt.get_cmap(cmap),
            background_color = bgcolor)

#show places
for name, location in places.items():
    coords = ((location[1] - rm.longs[0])/(rm.longs[1] - rm.longs[0]),
              (location[0] - rm.lats[0])/(rm.lats[1] - rm.lats[0]))   
    label = name
    ridges.plot(*coords, 'o', color="white",
            transform=ridges.transAxes,
            ms=3, zorder=5000)  
    ridges.text(coords[0]-0.005, coords[1]+0.015, label,
                fontproperties="Ubuntu", weight='ultralight',
                size=11,
                color="white",
                transform=ridges.transAxes,
                verticalalignment="bottom",
                bbox=dict(facecolor=bgcolor, edgecolor='white', linewidth=0, pad=1, alpha=0.6),
                zorder=len(values)+10)
    
#========
#title
plt.figtext(0.16,0.25, "Kasseler Berge".upper(), va='top', ha="left", fontsize=30, color="white", fontname="Ubuntu", 
            bbox=dict(facecolor=bgcolor, edgecolor='white', linewidth=0, pad=10, alpha=1),
            weight='ultralight')

#save
plt.savefig("kasseler-berge-red.png", bbox_inches='tight',dpi=300,  transparent=False, pad_inches=0)
#plt.savefig("kasseler-berge.svg", bbox_inches='tight', dpi=300, transparent=False, pad_inches=0)

plt.show()
```
