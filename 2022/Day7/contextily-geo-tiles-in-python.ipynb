```python
import contextily as cx
import geopandas as gpd
import rasterio
from rasterio.plot import show as rioshow
import matplotlib.pyplot as plt
```

```python
map_styles = [cx.providers.OpenStreetMap.Mapnik,
              cx.providers.CartoDB.Positron,
              cx.providers.CartoDB.Voyager,
              cx.providers.OpenTopoMap,
              cx.providers.Stamen.TonerLite,
              cx.providers.Stamen.Watercolor,
              cx.providers.Esri.WorldTerrain,
              cx.providers.Esri.OceanBasemap
             ]

titles = ["OpenStreetMap.Mapnik", "CartoDB.Positron", "CartoDB.Voyager", "OpenTopoMap", "Stamen.TonerLite",
         "Stamen.Watercolor", "Esri.WorldTerrain", "Esri.OceanBasemap"]
```

```python
fig, axs = plt.subplots(2,4, figsize=(15,10), layout="tight")

for i, source in enumerate(map_styles): 
    ax =  cx.Place("United Kingdom",  source=source).plot(ax=axs.flatten()[i], attribution="")
    ax.axis("off")
    ax.set_title(titles[i], fontsize=14)

fig.text(0.5,1,"CONTEXTILY - Context geo tiles in Python", ha="center", fontsize=20, fontweight="bold")

plt.savefig("context-geo-tiles-in-python.png", bbox_inches='tight', transparent=False, pad_inches=0.2)
plt.show()
```
