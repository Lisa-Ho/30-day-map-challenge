Using one of my most favourite python libraries: `ridge-map`

Checkout the [documentation](https://pypi.org/project/ridge-map/) for more details on how to get started. 

## Setup

```python
import matplotlib.pyplot as plt
from ridge_map import RidgeMap
```

```python
# set colormap
import matplotlib.colors as colors
from palettable.cartocolors.sequential import SunsetDark_7_r
cmap = colors.LinearSegmentedColormap.from_list('cmap',[colors.hex2color(c) for c in  SunsetDark_7_r.hex_colors[1:]])
cmap
```

```python
# custom fonts
from matplotlib import font_manager
plt.rcParams['font.family'] = ['FreeSans']
title_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\coolvetica rg.otf')
```

## Plot

```python
# get data for set coordinages
rm = RidgeMap((-10.898438,49.839465,2.373047,59.622063), font='FreeSans')
values = rm.get_elevation_data(num_lines=200, elevation_pts=200)
```

```python
# colours
textcolor= '#CDCDCD'
bgcolor = '#111111'

# setup figure
fig,ax = plt.subplots(figsize=(12, 12))
fig.set_facecolor(bgcolor)

# plot ridges
ridges = rm.plot_map(values=rm.preprocess(values=values, vertical_ratio=40, water_ntile=20, 
                                 lake_flatness=2), 
            ax=ax, kind='elevation', label=None,
            linewidth=1,
            line_color = plt.get_cmap(cmap),
            background_color = bgcolor)

# title
text = 'United Kingdom\n& Ireland'.upper()
plt.figtext(0.16, 0.92 , text, ha='left', color=textcolor, fontproperties=title_font, fontsize=32, )
plt.figtext(0.5, 0.05 , 'Data: NASA via ridge-map.py  |  Design: Lisa Hornung', fontsize=8, fontweight='regular', ha='center', color=textcolor)

# export
plt.savefig('uk-ridgemap-v2.png', dpi=120, bbox_inches='tight', pad_inches=1.4, facecolor=bgcolor)

plt.show()
```
